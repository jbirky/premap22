```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
rc('xtick', labelsize=20)
rc('ytick', labelsize=20)
rc('axes', facecolor='white')
```

```python
kep = pd.read_csv("../tables/kepler_sample_classified.csv")
tess = pd.read_csv("../tables/tess_feature_predicted_classifications.csv")

def plot_color_diagram(obj_id):
    plt.figure(figsize=[8,6])
    for ctype in set(kep['Class']):
        sel = kep[kep['Class'] == ctype]
        plt.scatter(sel['BP-RP'], sel['Gmag'], label=ctype)
    sel_tess = tess[tess['TIC'] == obj_id]
    plt.scatter(sel_tess['BP-RP'], sel_tess['Gmag'], color='k', s=20, label=obj_id)
    plt.legend(loc='best', fontsize=18)

def plot_diagnostic_diagram(obj_id):
    plt.figure(figsize=[10,8])
    for ctype in set(kep['Class']):
        sel = kep[kep['Class'] == ctype]
        plt.scatter(sel['period'], sel['morph'], label=ctype)
    sel_tess = tess[tess['TIC'] == obj_id]
    plt.scatter(sel_tess['period_x'], sel_tess['morph_x'], color='k', s=20, label=obj_id)
    plt.xscale('log')
    plt.ylim(0, 1)
    plt.legend(loc='best', fontsize=18)
```

```python
plot_diagnostic_diagram(25078924)
```
